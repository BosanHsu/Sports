```python
import pandas as pd
import numpy as np
```

```python
NBAmod = pd.read_excel("Assignment Data/Week 4/NBA prediction model (Assignment).xlsx")
NBAmod
```

### Step 1: Data preparation

```python
# 1-1 Define variables for the probabilities of a home win and away win associated with bookmaker odds

NBAmod["bookmaker_hwodds"] = (1/NBAmod["hwinodds"]) / ( (1/NBAmod["hwinodds"]) + (1/NBAmod["hloseodds"]) )
NBAmod["bookmaker_awodds"] = 1 - NBAmod["bookmaker_hwodds"]
NBAmod
```

```python
# 1-2 Define a dummy variable = 1 if the home team loses, and zero otherwise
NBAmod["hlose"] = 1 - NBAmod["hwin"]
```

```python
# 1-3 Define a variable equal to H if the home team wins and A if the visiting team wins
NBAmod["win_team"] = np.where( NBAmod["hwin"] == 1, "H", "A" )
```

```python
# 1-4 Define a variable equal to H if the home team win probability is greater than 0.5 according to the bookmaker odds and A otherwise
NBAmod["pred_hwin"] = np.where( NBAmod["bookmaker_hwodds"] > 0.5, 1, 0 )
NBAmod
```

```python
# 1-5 Define a variable equal to the logarithm of the ratio of the home team salaries to the visiting team salaries.
NBAmod["log_salary_ratio"] = np.log( NBAmod["hteamsal"]/NBAmod["opposal"] )
NBAmod
```

### Step 2: Estimate a logit model of home time wins depending on the log salary ratio, using the data for calendar year 2018 as the “training data”.

```python
# 2-1 Define a subset for the calendar year 2018 data 
NBA2018 = NBAmod[ NBAmod["year"] == 2018 ]
NBA2018
```

```python
# 2-2 Import the logistic regression package
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# 2-3 Run the logistic regression of hwin on the log salary ratio
NBA2018_logit = smf.glm(formula = " hwin ~ log_salary_ratio ", data = NBA2018, family = sm.families.Binomial()).fit()
NBA2018_logit.summary()
```

### Step 3: Define the predicted probabilities and the predicted results, using the entire data set

```python
# 3-1 The predicted probability of home win can be defined using the formula:
# 1/(1+1/(exp(b0 + b1 (logsalaryratio))))
# where b0 is the constant (the intercept) in the logistic regression and b1 is the coefficient for logsalaryratio.

fittedProbs = NBA2018_logit.predict()
NBA2018['pred_hwin_logit'] = 1/(1+np.exp(-(0.4452 + 1.1216*NBA2018['log_salary_ratio'])))
NBA2018
```

```python
# 3-2 Based on the predicted probability,
# define the predicted result H as the outcome where the predicted home win probability is greater than 0.5, and A otherwise.

NBA2018['win_team_logit'] = np.where( NBA2018['pred_hwin_logit'] > 0.5, "H", "A" )
NBA2018
```

### Step 4: For games played in 2019, compare the bookmaker probabilities and model probabilities in terms of the mean number of successfully predicted outcomes and the Brier scores.

```python
# 4-1 Define the subset of games played in calendar year 2019
NBA2019 = NBAmod[ NBAmod["year"] == 2019 ]
```

```python
# 4-2 Define a dummy variable equal to 1 when the bookmaker result prediction is correct,
# and zero otherwise. Define the equivalent variable for the logit model prediction.
NBA2019["bookmaker_correct"] = np.where( NBA2019['pred_hwin'] == NBA2019['hwin'], 1, 0 )
```

```python
NBA2019['pred_hwin_logit'] = 1/(1+np.exp(-(0.4452 + 1.1216*NBA2019['log_salary_ratio'])))
NBA2019['win_team_logit'] = np.where( NBA2019['pred_hwin_logit'] > 0.5, "H", "A" )
NBA2019['hwin_logit'] = np.where( NBA2019['pred_hwin_logit'] > 0.5, 1, 0 )
NBA2019["logit_correct"] = np.where( NBA2019['win_team_logit'] == NBA2019['win_team'], 1, 0 )
```

```python
# 4-3 Calculate the means for each of these variables
NBA2019.describe()
```

```python
# 4-4 Define the Brier score for the bookmaker probabilities and the Brier score for the logit model probabilities
book_brier = (1/688)*((NBA2019["pred_hwin"] - NBA2019["hwin"])**2 + ((1-NBA2019["pred_hwin"]) - (1-NBA2019["hwin"]))**2).sum()
book_brier
```

```python
logit_brier = (1/688)*((NBA2019["pred_hwin_logit"] - NBA2019["hwin"])**2 + ((1-NBA2019["pred_hwin_logit"]) - (1-NBA2019["hwin"]))**2).sum()
logit_brier
```
