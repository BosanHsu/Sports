Calculate the implied win probability based on the bookmaker odds

```python
# win odds
# (a) taking the reciprocal of the odds
# (b) dividing by the sum of all probabilities
# (because of the overround we have to scale the probabilities so that they add up to 100%).
NBA19['winprob']= 1/(NBA19['winodds'])/(1/(NBA19['winodds'])+ 1/(NBA19['loseodds']))
NBA19
```

Drop from the data all games for which home = 0
