# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("../Data/Game logs 1999-2003.xlsx")
Teams.describe()
```

```python
Teams.dtypes
```

```python
# Load the “Game logs” data file,
# create a dummy variable for home wins (1 or 0) and visitor wins, and create the ‘year’ variable.
Teams["home_win"] = np.where( Teams["home_score"] > Teams["visitor_score"], 1, 0 )
Teams["visitor_win"] = np.where( Teams["home_score"] < Teams["visitor_score"], 1, 0 )

Teams["date"] = pd.to_datetime(Teams["date"], format='%Y%m%d')
Teams["years"] = Teams['date'].dt.year

Teams
```

```python
# Create a dataframe that sums the statistics you need by season (year) and by home team.
# Rename the column for the home team as ‘Team’.

Home_Teams = Teams.groupby(["years","home"]).sum().reset_index()
Home_Teams = Home_Teams.rename(columns={'home': 'Team'})
Home_Teams.head()
```

```python
# Create a dataframe that sums the statistics you need by season (year) and by visiting team.
# Rename the column for the visiting team as ‘Team’.
Away_Teams = Teams.groupby(["years","visitor"]).sum().reset_index()
Away_Teams = Away_Teams.rename(columns={'visitor': 'Team'})
Away_Teams.head()
```

```python
# Merge the two dataframes on ‘year’ and ‘Team’ and create a variable for the total number of wins in the season for each team 
Merged_Teams = pd.merge(Away_Teams, Home_Teams, on=['years', 'Team'])
Merged_Teams.head()
```

```python
Merged_Teams["Total_Win"] = Merged_Teams["home_win_y"] + Merged_Teams["visitor_win_x"]
Merged_Teams.head()
```

# II. Calculating Performance Metrics

```python
# Create batting average for and batting average against as the sum the relevant statistics for each team
# as home team and visiting team.
Each_Team = pd.DataFrame()
Each_Team["home_avg"] = Home_Teams.groupby("Team")["home_h"].sum()/Home_Teams.groupby("Team")["home_ab"].sum()
Each_Team["home_avg_allowed"] = Home_Teams.groupby("Team")["visitor_h"].sum()/Home_Teams.groupby("Team")["visitor_ab"].sum()
Each_Team["away_avg_allowed"] = Away_Teams.groupby("Team")["home_h"].sum()/Away_Teams.groupby("Team")["home_ab"].sum()
Each_Team["away_avg"] = Away_Teams.groupby("Team")["visitor_h"].sum()/Away_Teams.groupby("Team")["visitor_ab"].sum()
Each_Team
```

```python
# Create the win percentage variable for each team by generating a count of games played
# and dividing total wins by total games played.
Each_Team["total_games"] = Teams.groupby("home")["date"].count() + Teams.groupby("visitor")["date"].count()
Each_Team["total_win"] = Home_Teams.groupby("Team")["home_win"].sum() + Home_Teams.groupby("Team")["home_win"].sum()
Each_Team["total_wpct"] = Each_Team["total_win"]/Each_Team["total_games"]
Each_Team
```

```python
# Using a similar process to part 2), calculate each team’s home and away win percentage separately for each season
Each_Season = pd.DataFrame()
Each_Season["home_wpct"] = Home_Teams.groupby(["years","Team"])["home_win"].sum()/Teams.groupby(["years","home"])["date"].count()
Each_Season["away_wpct"] = Away_Teams.groupby(["years","Team"])["visitor_win"].sum()/Teams.groupby(["years","visitor"])["date"].count()
Each_Season["total_wpct"] = ( Home_Teams.groupby(["years","Team"])["home_win"].sum() + Away_Teams.groupby(["years","Team"])["visitor_win"].sum()) / ( Teams.groupby(["years","home"])["date"].count() + Teams.groupby(["years","visitor"])["date"].count() )
Each_Season.head()
```

```python
# Merge each team’s home and away win percentage for each season into one data frame.
# Calculate WPCT diff defined as home win percentage – away win percentage for every team in each season.
Each_Season["wpct_diff"] = Each_Season["home_wpct"] - Each_Season["away_wpct"]
Each_Season.head()
```

Question 1
Find the team with the maximum single season WPCT diff in the dataframe. What was this team’s away winning percentage for the season?

```python
Each_Season[ Each_Season["wpct_diff"] == Each_Season["wpct_diff"].max()]
```

Question 2
Create a variable AVG Diff = Batting Average For – Batting Average Against. What was the minimum (most negative) single season team AVG Diff in the dataframe?

```python
Each_Season["h"] = Home_Teams.groupby(["years","Team"])["home_h"].sum() + Away_Teams.groupby(["years","Team"])["visitor_h"].sum()
Each_Season["ab"] = Home_Teams.groupby(["years","Team"])["home_ab"].sum() + Away_Teams.groupby(["years","Team"])["visitor_ab"].sum()
Each_Season["avg"] = Each_Season["h"]/Each_Season["ab"]

Each_Season["h_opp"] = Home_Teams.groupby(["years","Team"])["visitor_h"].sum() + Away_Teams.groupby(["years","Team"])["home_h"].sum()
Each_Season["ab_opp"] = Home_Teams.groupby(["years","Team"])["visitor_ab"].sum() + Away_Teams.groupby(["years","Team"])["home_ab"].sum()
Each_Season["avg_opp"] = Each_Season["h_opp"]/Each_Season["ab_opp"]

Each_Season["avg_diff"] = Each_Season["avg"] - Each_Season["avg_opp"]

Each_Season[ Each_Season["avg_diff"] == Each_Season["avg_diff"].min()]
```

# III. Running Regressions

```python
# Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against.
import statsmodels.formula.api as smf
lm = smf.ols(formula = 'total_wpct ~ avg + avg_opp', data = Each_Season).fit()
lm.summary()
```

```python
# Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against,
# constraining the coefficient on each variable to be equal and opposite.
lm_constrained = smf.ols(formula = 'total_wpct ~ avg_diff', data = Each_Season).fit()
lm_constrained.summary()
```
